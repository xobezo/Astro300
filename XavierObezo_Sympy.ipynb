## Part I

$$ \Large {\displaystyle f(x)=3e^{-{\frac {x^{2}}{8}}}} \sin(x/3)$$

* Find the first four terms of the Taylor expansion of the above equation (at x = 0)
* Make a plot of the function
* Plot size 10 in x 4 in
* X limts -5, 5
* Y limits -2, 2
* On the same plot:
  * Over-plot the 1-term Taylor expansion using a different color
  * Over-plot the 2-term Taylor expansion using a different color
  * Over-plot the 3-term Taylor expansion using a different color
  * Over-plot the 4-term Taylor expansion using a different color

```python
#Define Function
x= sp.symbols('x')
f = 3*sp.exp(-x**2/8) * sp.sin(x/3)
#First 4-terms of Taylor series
taylor = sp.series(f,x,x0=0,n=8)
taylor.removeO()
```

```python
#Graph
# Equation
# Main Equation
np_x = np.linspace(-5,5 ,200)
fNumpy = 3*np.e**(-np_x**2/8) * np.sin(np_x/3)
# First Order
f_1 = np_x
# Second Order
f_2 = f_1 - 31*np_x**3/216
# Third Order
f_3 = f_2 + 1591*np_x**5/155520
# Fourth Order
f_4 = f_3 - 113653*np_x**7/235146240

#Basic Definitions
fig,ax = plt.subplots(1,1)
fig.set_size_inches(10,4)
fig.tight_layout()


#Title
plt.title('Graph')

#Limits
ax.set_ylim(-2,2)
ax.set_xlim(-5,5)

#Labels
ax.set_xlabel("This is X")
ax.set_ylabel("This is Y")

# For the approximation line size
appSizes = 3

#Graphs
# Main Function
ax.plot(np_x, fNumpy, color='r', marker='None', linestyle='-', linewidth=4,label="Function");
# First Order
ax.plot(np_x, f_1, color='g', marker='None', linestyle='--', linewidth=appSizes,label="One Term");
# Second Order
ax.plot(np_x, f_2, color='b', marker='None', linestyle='--', linewidth=appSizes,label="Two Terms");
# Third Order
ax.plot(np_x, f_3, color='orange', marker='None', linestyle='--', linewidth=appSizes,label="Three Terms");
# Fourth Order
ax.plot(np_x, f_4, color='black', marker='None', linestyle='--', linewidth=appSizes,label="Four Terms");
# Add Legend
ax.legend(loc=0,shadow=True);

```

## Part II

$$\Large {\displaystyle g(x)=x^{2} - 0.75}$$


#### What are the value(s) for `x` where f(x) = g(x)

```python
g = x**2 - 0.75
np_g = np_x**2 - 0.75

#Basic Definitions
fig,ax = plt.subplots(1,1)
fig.set_size_inches(10,4)
fig.tight_layout()

#Limits
ax.set_ylim(-2,2)
ax.set_xlim(-2,2)

#Graphs
# Main Function
ax.plot(np_x, fNumpy, color='r', marker='None', linestyle='-', linewidth=4,label="Function");
ax.plot(np_x, np_g, color='r', marker='None', linestyle='-', linewidth=4,label="Function");

#Title
plt.title('Finding Guess')

guesses = [-0.5,1.2]

for val in guesses:
    result = sp.nsolve(f - g, x, val)
    print("They meet at x={0:.5f}.".format(result))
```
